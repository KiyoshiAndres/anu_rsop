```python
import persistence_functions as pf
```

```python
import trimesh

# Load the .obj file
mesh = trimesh.load('triangulated_torus_2.obj')

# Print basic information about the mesh
print(mesh)
```

```python
torus_edges = mesh.edges_unique
torus_vertices = mesh.vertices
torus_vertices_normals = mesh.vertex_normals
```

```python
direction = [0,1,1]
```

```python
pre_formatted_vertices = []
for vertex, normal in zip(torus_vertices, torus_vertices_normals):
    height = pf.height_of_vertex(direction, vertex)
    new_vertex = [vertex, height, normal]
    pre_formatted_vertices.append(new_vertex)
```

```python
formatted_vertices = pf.format_vertices(pre_formatted_vertices)
```

```python
pre_formatted_edges = []
for edge in torus_edges:
    x,y = edge
    pre_formatted_edges.append([[x + 1, y + 1], [1,1,1]])
```

```python
formatted_edges = pf.format_edges(formatted_vertices, pre_formatted_edges)
```

```python
uf = pf.UnionFind(formatted_vertices)
```

```python
for edge in formatted_edges:
    x ,y = edge['vertices']
    uf.union(x - 1, y - 1)

# After processing all edges, we collect the connected components.
components = {}
for node in range(len(formatted_vertices)):
    root = uf.find(node)  # This finds the representative of the component containing 'node'
    if root not in components:
        components[root] = []
    components[root].append(node)

print("Final connected components:")
for root, comp in components.items():
    print(f"Component with root {root}: {comp}")
```
